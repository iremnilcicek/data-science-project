```python


# Veriyi gruplama ve platform_gender_counts oluşturma
platform_gender_counts = df.groupby(["Gender", "PlatformWorkedWith"]).size().reset_index(name="Count")

# Cinsiyetleri al
genders = platform_gender_counts['Gender'].unique()

# Her cinsiyet için pasta grafiği oluşturma
for gender in genders:
    data = platform_gender_counts[platform_gender_counts['Gender'] == gender]  # Düzeltildi
    plt.figure(figsize=(10, 10))
    plt.pie(
        data['Count'], 
        labels=data['PlatformWorkedWith'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=plt.cm.tab20.colors
    )
    plt.title(f"{gender} Çalışanlarının Kullandığı Platformlar")
    plt.show()
```

```python
# Veriyi hazırlama: Cinsiyet ve platforma göre kişi sayısını hesapla
platform_gender_counts = df.groupby(['Gender', 'PlatformWorkedWith']).size().reset_index(name='Count')

# Benzersiz cinsiyetleri al
genders = platform_gender_counts['Gender'].unique()

# Grafik düzeni (örneğin, 1 satır ve 3 sütun)
fig, axes = plt.subplots(1, len(genders), figsize=(18, 6))

# Her cinsiyet için ayrı bir pasta grafiği oluşturma
for i, gender in enumerate(genders):
    data = platform_gender_counts[platform_gender_counts['Gender'] == gender]
    axes[i].pie(
        data['Count'], 
        labels=data['PlatformWorkedWith'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=plt.cm.tab20.colors
    )
    axes[i].set_title(f"{gender} Çalışanlarının Kullandığı Platformlar")

# Tüm grafikleri düzenleme
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Veriyi hazırlama: Cinsiyet ve platforma göre kişi sayısını hesapla
platform_gender_counts = df.groupby(['Gender', 'PlatformWorkedWith']).size().reset_index(name='Count')

# Treemap grafiği oluşturma
fig = px.treemap(platform_gender_counts, 
                 path=['Gender', 'PlatformWorkedWith'], 
                 values='Count', 
                 color='Count', 
                 hover_data=['Count'],
                 title='Cinsiyete Göre Çalışanların Kullandığı Platformlar')

# Grafiği gösterme
fig.show()
```

```python
# Veriyi gruplama
grouped_data = df.groupby(['Gender', 'LanguageWorkedWith'])['Age'].mean().reset_index()

# Pivot table oluşturma (stacked bar için uygun hale getirme)
pivot_data = grouped_data.pivot(index='LanguageWorkedWith', columns='Gender', values='Age')

# Pivot tablodaki boş değerleri doldurma
pivot_data = pivot_data.fillna(0)

# Stacked bar chart oluşturma
pivot_data.plot(kind='bar', stacked=True, figsize=(16, 9), colormap='viridis')

# Grafik ayarları
plt.xticks(rotation=45)
plt.ylabel('Ortalama Yaş')
plt.title('Programlama Dillerine Göre Ortalama Yaş (Cinsiyet Bazlı)', fontsize=14)
plt.legend(title='Cinsiyet')
plt.tight_layout()
plt.show()
```

```python
# Veri Yükleme
file_path = '/Users/iremnilcicek/Desktop/Demographics and technologies normalised Data.csv'
data = pd.read_csv(file_path)

# Eksik Verileri İnceleme ve Temizleme
data = data.dropna(subset=['Age', 'Gender', 'LanguageWorkedWith'])  # Eksik yaş, cinsiyet veya programlama dili olanları çıkarıyoruz

# Gerekli Sütunları Seçme
data = data[['Gender', 'Age', 'LanguageWorkedWith']]

# Programlama Dilleri Sütununu Ayrıştırma
data = data.assign(LanguageWorkedWith=data['LanguageWorkedWith'].str.split(';'))  # Dilleri ayırıyoruz

data = data.explode('LanguageWorkedWith')  # Her dili ayrı bir satıra dönüştürüyoruz

# Cinsiyete ve Programlama Diline Göre Yaş Ortalaması
grouped_data = data.groupby(['Gender', 'LanguageWorkedWith'])['Age'].mean().reset_index()

# 3D Görselleştirme
fig = px.scatter_3d(
    grouped_data,
     x='LanguageWorkedWith',
    y='Age',
    z='Gender',
    color='Gender',
    title='Programming Languages, Gender, and Average Age Distribution',
    labels={
        'LanguageWorkedWith': 'Programming Language',
        'Age': 'Average Age',
        'Gender': 'Gender'
    },
    template='plotly_dark'
)

# Grafik Ayarları
fig.update_traces(marker=dict(size=10))  # Nokta boyutları
fig.update_layout(scene=dict(
    xaxis_title='Programming Language',
    yaxis_title='Average Age',
    zaxis_title='Gender'
))

fig.show()
```
